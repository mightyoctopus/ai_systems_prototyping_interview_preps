# Auth

## Write 2 functions 
- hash_password
- verify_password

```python
import bcrypt
```

```python
def hash_password(password):
    encoded = password.encode("utf-8")
    hashed = bcrypt.hashpw(encoded, bcrypt.gensalt())

    return hashed.decode("utf-8")
```

```python
def verify_password(user_input, hashed_pw):

    return bcrypt.checkpw(
        user_input.encode("utf-8"), 
        hashed_pw.encode("utf-8")
    )
```

```python
verify_password("my_secure_passsword", hash_password("my_secure_passsword"))
```
