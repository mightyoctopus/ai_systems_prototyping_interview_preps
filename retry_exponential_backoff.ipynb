```python
import time
```

```python
def call_api(): 
    """
    Simulates an API failure
    """

    raise Exception("fail")
```

```python
def call_with_retry():
    for attempt in range(5):
        try: 
            return call_api()
        except:
            wait = 2 ** attempt
            print(f"Wait time: ")
            time.sleep(wait)

    return "fallback"
```

```python
call_with_retry()
```
