```python
cache = {}
```

```python
def normalize(query):
    return query.strip().lower()
```

```python
def llm(query):
    return f"Example answer generated by llm for {query}"
```

```python
def get_cached_asnwer(query):

    key = normalize(query)

    if key in cache: 
        return cache[key]

    result = llm(query)
    cache[key] = result

    return result
```

```python
get_cached_asnwer("how's the weather today?")
```
